# Presentation Visuals

## Set-up

### Imports

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import warnings
from visualization_utils import (plot_avg_total_investment_periods, 
                                 plot_joint_distribution, 
                                 update_facet_grid)
```

### Load Data

```python
mgpgg_df = pd.read_csv('/Users/aaronberman/Desktop/Live_Visuals/src/mgpgg_live_visuals_df.csv', index_col=0)
```

### Constants

```python
plt.rcParams["font.family"] = "Times New Roman"
warnings.filterwarnings("ignore")
palette = {"Multi-Split": "#FF6347", "Multi-Shared": "#4682B4"}  # Tomato and Steel Blue
mgpgg_only_df = mgpgg_df[(mgpgg_df['treatment'].isin(['Multi-Split', 'Multi-Shared']))]
```

## Plots

### Average Total Investment Across Periods

```python
plot_avg_total_investment_periods(mgpgg_df)
```

### Joint and Marginal Distribution Plots (Interactive)

```python
period_slider_joint = widgets.IntSlider(min=1, max=20, step=1, description='Period:', value=1)
widgets.interactive(lambda period: plot_joint_distribution(mgpgg_only_df, period, palette), period=period_slider_joint)
```

### Joint Distribution Plot – FacetGrid (Interactive)

```python
def interactive_facet_grid(period):
    update_facet_grid(data=mgpgg_df, period=period, palette=palette)

period_slider_facet = widgets.IntSlider(min=1, max=20, step=1, description='Period:', value=1)
widgets.interactive(interactive_facet_grid, period=period_slider_facet)
```
